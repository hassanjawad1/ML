```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df["fare"] = df["fare"].fillna(df["fare"].median())
```

```python
from sklearn.preprocessing import KBinsDiscretizer
age_discretizer = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='quantile')
df['age_bin']= age_discretizer.fit_transform(df[['age']])
df.head()
```

```python
sns.histplot(df, x="age" , hue= 'age_bin')
```

```python
# using pandas now
bins= [0,10,18,30,50,80]
labels= ['child','teen','young','adult','old']
df['age_bin2']= pd.cut(df['age'], bins=bins, labels=labels)
sns.histplot(df, x="age" , hue= 'age_bin2')
```
